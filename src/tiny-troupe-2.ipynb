```python
general_context = f"""
            We are performing a survey of citizens of Europe during World War I. 
            The sample must include representative people of the European population around 1914-1921. 
            The sample must contain a diverse group:
            - from male to female
            - from young to old
            - from wealthy aristocrats to poor folks
            - from lowly educated to highly educated
            Most importantly, they also must hold a distinct political belief:
            - supports the axis powers (including Germany) vs. allied powers (including France)
            - holds democratic beliefs vs. authocratic ideas
            - believes in capitalism vs socialism
              """

general_context_2 = f"""
            We are performing a survey of citizens of Europe during World War I. 
            The sample must include representative people of the German population around 1914-1921. 
            The sample must contain a diverse group:
            - from male to female
            - from young to old
            - from wealthy aristocrats to poor folks
            - from lowly educated to highly educated
            Most importantly, they ALL must be STRONG supporters of the Germans during World War I.
              """
```

```python
first_interview_prompt = f"""Read the following German article during World War I writen in Belgium. \n\n 
{text} \n\n
How does it make you feel? How does it inform your views during the ongoing war?"""
extractor_objective_1 = "Give me a summary of the main ideas or themes from the people."
extractor_objective_2 = "Give me a structured data to make a word cloud on the main ideas or themes. Make sure to consolidate similar words or themes if possible."
```

```python
second_interview_prompt = f"""Now read the French response to the previous article, also written during World War I. 
{text2} \n\n
How does this change how you feel? How does it change your views during the ongoing war?"""
extractor_objective_3 = "Give me a summary of the main ideas or themes from the people. After reading the French response, how did people change their opinion based on 1. wealth 2. education 3. gender 4. politics?"
extractor_objective_4 = "Give me a structured data to make a word cloud on the main ideas or themes after they read the French response. Make sure to consolidate similar words or themes if possible."
```

```python
factory = TinyPersonFactory(general_context)
factory2 = TinyPersonFactory(general_context_2)
```

```python
people = factory.generate_people(15)
germans = factory2.generate_people(15)
```

```python
extractor = ResultsExtractor()
```

```python
# first interview
for p in people:
    p.listen_and_act(first_interview_prompt)

for g in germans:
    g.listen_and_act(first_interview_prompt)
```

```python
results_1_people = extractor.extract_results_from_agents(people, extraction_objective=extractor_objective_1)
results_1_germans = extractor.extract_results_from_agents(germans, extraction_objective=extractor_objective_1)
results_2_people = extractor.extract_results_from_agents(people, extraction_objective=extractor_objective_2)
results_2_germans = extractor.extract_results_from_agents(germans, extraction_objective=extractor_objective_2)
```

```python
for p in people:
    p.listen_and_act(second_interview_prompt)

for g in germans:
    g.listen_and_act(second_interview_prompt)
```

```python
results_3_people = extractor.extract_results_from_agents(people, extraction_objective=extractor_objective_3)
results_3_germans = extractor.extract_results_from_agents(germans, extraction_objective=extractor_objective_3)
results_4_people = extractor.extract_results_from_agents(people, extraction_objective=extractor_objective_4)
results_5_germans = extractor.extract_results_from_agents(germans, extraction_objective=extractor_objective_4)
```
